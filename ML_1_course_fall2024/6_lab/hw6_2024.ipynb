Ученые придумали новый тест для определения одной очень известной болезни, который показывает поразительные свойства:
* Вероятность того, что тест покажет положительный результат, если человек и правда болен, составляет 99%;
* Вероятность того, что тест покажет положительный результат, если человек на самом деле не болен, составляет 3%.

Посчитайте вероятность того, что человек реально болен, если тест показал положительный результат, с помощью формулы Байеса. Также известно, что в среднем на 100 случайных человек приходится двое болеющих.

Пусть $P(Б)$ - вероятность что человек болен, а $P(З)$ - вероятность что человек здоров

$P(Б) = 2/100$

$P(З) = 1 - P(Б) = 98/100$

Пусть $P(П | Б)$ - вероятность что получен положительный результат теста при условии что человек болен
а $P(O | Б)$ - вероятность получить отрицательный тест, если человек болен

$P(П | Б) = 0.99$

$P(О | Б) =1- P(П | Б)= 0.01$

Пусть $P(П | З)$ - вероятность получить положительный результат, если человек здоров, а $P(О | З)$ - вероятность получить отризательный результат, если человек здоров.

$P(П | З) = 0.03$

$P(О | З) = 1 - P(П | З) = 0.97$

Нам нужно найти $P(Б | П) = \dfrac{P(П | Б)P(Б)}{P(П)}$

Для этого найдем $P(П)$ по формуле полной вероятности:
$P(П) = P(П | Б)*P(Б) + P(П | З)*P(З)

$P(Б | П) = \dfrac{P(П | Б) \times P(Б)}{P(П | Б) \times P(Б) + P(П | З) \times P(З)} =0.4 $

### Задача 2 (выполнено). 

Рассмотрим два случая.

**a).** Пусть $X_1, \dots, X_n$ &mdash; выборка из распределения $\mathcal{N}(\theta, 1)$. 

**b).** Пусть $X_1, \dots, X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \theta^{-1})$. 

В каждом случае подберите сопряженное семейство распределений и найдите соответствующее апостериорное распределение параметра $\theta$, а также его математическое ожидание.

Подумайте, какое распределение является сопряженным в случае если неизвестны оба параметра нормального распределения.


$X_1, \dots, X_n \sim \mathcal{N}(\theta, 1)$

Правдоподобие:
$$
L(\theta | X) \propto \exp\left(-\frac{1}{2} \sum_{i=1}^n (X_i - \theta)^2\right).
$$

Сопряжённое распределение:

Для нормального распределения с известной дисперсией $1$, сопряжённым распределением параметра $\theta$ является нормальное распределение:
$$
\theta \sim \mathcal{N}(\mu_0, \tau_0^2),
$$
где $\mu_0$ и $\tau_0^2$ — параметры априорного распределения.

Апостериорное распределение:

После получения выборки, апостериорное распределение параметра $\theta$ также является нормальным:
$$
\theta | X \sim \mathcal{N}(\mu_n, \tau_n^2),
$$
где:
$$
\tau_n^2 = \frac{1}{\frac{1}{\tau_0^2} + n}, \quad
\mu_n = \tau_n^2 \left(\frac{\mu_0}{\tau_0^2} + n\bar{X}\right).
$$

Математическое ожидание:
$$
\mathbb{E}[\theta | X] = \mu_n = \frac{\mu_0 / \tau_0^2 + n\bar{X}}{1 / \tau_0^2 + n}.
$$

---

$X_1, \dots, X_n \sim \mathcal{N}(0, \theta^{-1})$

Правдоподобие:
$$
L(\theta | X) \propto \theta^{n/2} \exp\left(-\frac{\theta}{2} \sum_{i=1}^n X_i^2\right).
$$

Сопряжённое распределение:

Для нормального распределения с неизвестной обратной дисперсией ($\theta$), сопряжённым распределением является гамма-распределение:
$$
\theta \sim \text{Gamma}(\alpha_0, \beta_0),
$$
где $\alpha_0$ и $\beta_0$ — априорные параметры.

Апостериорное распределение:

После получения выборки, апостериорное распределение $\theta$ остаётся гамма-распределением:
$$
\theta | X \sim \text{Gamma}(\alpha_n, \beta_n),
$$
где:
$$
\alpha_n = \alpha_0 + \frac{n}{2}, \quad
\beta_n = \beta_0 + \frac{1}{2} \sum_{i=1}^n X_i^2.
$$

Математическое ожидание:
$$
\mathbb{E}[\theta | X] = \frac{\alpha_n}{\beta_n} = \frac{\alpha_0 + n/2}{\beta_0 + \frac{1}{2} \sum_{i=1}^n X_i^2}.
$$

---

Если оба параметра нормального распределения неизвестны, сопряжённым семейством является комбинация:
- Нормального распределения для $\mu$,
- Обратного гамма-распределения для $\sigma^2$.

Апостериорное распределение:
$$
\mu | \sigma^2 \sim \mathcal{N}(\mu_n, \sigma^2 / \kappa_n), \quad \sigma^2 \sim \text{InvGamma}(\alpha_n, \beta_n).
$$

Математическое ожидание:
$$
\mathbb{E}[\mu | X] = \mu_n, \quad \mathbb{E}[\sigma^2 | X] = \frac{\beta_n}{\alpha_n - 1}.
$$

```python
def eval_c(sample_norm, cauchy_sample_size=100000):
    '''
    Функция для оценки константы в знаменателе формулы Байеса.
    
    :param sample_norm: выборка из нормального распределения.
    :param cauchy_sample_size: размер выборки для метода Монте-Карло
                        
    :returns: оценка константы
    '''
    
    sample_cauchy = sps.cauchy(loc=24).rvs(
        size=cauchy_sample_size
    ).reshape(1, -1)
    sample_norm = sample_norm.reshape(-1, 1)
    
    return np.mean(
        sps.norm(loc=sample_cauchy, scale=sigma).pdf(sample_norm).prod(axis=0)
    )
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as sps


sigma = 11. / 3
cauchy_loc = 24  
batch = 20

iter = np.round(df["Alcalinity of ash"].size / batch) # это не очень хорошо так округлять но в данном случае норм
start_idx = 0
end_idx = batch - 1 


sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))

for i in range(int(iter)):
    sample_norm = df["Alcalinity of ash"].loc[:end_idx].values
    c = eval_c(df["Alcalinity of ash"].loc[:end_idx].values)  
    end_idx += batch
    grid = np.linspace(-3 + sample_norm.mean(), 3 + sample_norm.mean(), 500000)

    
    density = np.exp(-((sample_norm.reshape(-1, 1) - grid) / sigma) ** 2 / 2).prod(axis=0) * \
              sps.cauchy(loc=cauchy_loc).pdf(grid) / c * \
              np.power(1 / np.sqrt(2 * np.pi) / sigma, len(sample_norm))

    
    plt.plot(grid, density, label=f"кол-во объектов={len(sample_norm)}", lw=2.5, color=(1 - i / iter, i / iter, 0))


plt.title('Изменение апостериорного распределения с добавлением данных')
plt.xlabel("Параметр $\\theta$")
plt.legend(fontsize=10, loc='upper right')
plt.show()
```

**Вывод:**

С маленькой выборкой получается большой разброс, поэтому левые распределения сильнее приплюснуты. Однако с ростом количества объектов оценка параметра сходится к выборочному среднему, распрделение "вытягивается"

Некоторая ML-модель имеет один параметр $\theta$, который обучается посредством *минимизации* функции
$$\mathcal{L}(\theta) = \mathsf{E} \exp\left(-\frac{\xi^2\sqrt{\eta}}{1+\theta^2}\right),$$
где $\xi$ имеет стандартное нормальное распределение, а $\eta$ &mdash; пуассоновское распределение с параметром 5 и не зависит от $\eta$.


Оптимизируйте эту функцию, используя стохастический градиентный спуск. Необходимо записать формулу шага и реализовать процедуру оптимизации.

```python
from sympy import symbols, sqrt, exp, diff
from sympy import latex
from IPython.display import display, Math
import numpy as np

theta, xi, eta = symbols('theta xi eta')
L = exp(- sqrt(eta)*xi**2 / (theta**2+1))

#возьмём производную
dL_dtheta = diff(L, theta)
print("Производная:")
dL_dtheta
```

```python
from sympy import symbols, sqrt, exp, diff, lambdify
import matplotlib.pyplot as plt

theta_sym, xi_sym, eta_sym = symbols('theta xi eta')
grad_fn = lambdify((theta_sym, xi_sym, eta_sym), dL_dtheta, 'numpy')
thetas = []
iterations = 10000
xi_samples = np.random.normal(0, 1, iterations)  
eta_samples = np.random.poisson(5, iterations) 
for i in range(iterations):
    theta = 1
    epoch = 15
    for j in range(epoch):
        grad = grad_fn(theta, xi_samples[i], eta_samples[i])
        theta -= grad*10
    thetas.append(theta)

cumu_mean = np.cumsum(thetas) / np.arange(1, len(thetas) + 1)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(thetas) + 1), cumu_mean, label="20 эпох")
plt.xlabel("итерации", fontsize=12)
plt.ylabel("накопленное среднее", fontsize=12)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```
